```python
plt.figure(figsize = (20,5))
ax = sns.countplot(data= df, x="SepalLengthCm", hue = "Species") 
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

SepalLengthCm has 2 max count of "Iris-setosa" .

```python
plt.figure(figsize = (20,5))
ax = sns.countplot(data= df, x="SepalWidthCm", hue= "Species")  
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

 SepalWidthCm has max count of "Iris-virginica" & then "Iris-setosa"

```python
plt.figure(figsize = (20,5))
ax = sns.countplot(data= df, x="PetalLengthCm", hue = "Species") 
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

PetalLengthCm has max count of "Iris-setosa".

```python
plt.figure(figsize = (20,5))
ax = sns.countplot(data= df, x="PetalWidthCm", hue= "Species") 
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

PetalWidthCm has max count of "Iris-setosa".

Mostly max. values are related to "Iris-setosa"

```python
gb = df.groupby("Species").agg({"SepalLengthCm":'mean',"SepalWidthCm":'mean',"PetalLengthCm":'mean',"PetalWidthCm":'mean'})
gb
sns.heatmap(gb,annot = True)
plt.show()
```
